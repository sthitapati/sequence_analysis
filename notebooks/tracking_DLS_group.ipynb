```python
# write a function to read the dates of the video from the h5 filenames (can be one name or a list of names)
# e.g. camera12024-02-02T12_19_31DLC_resnet50_seq_8_keypoints_thethredFeb27shuffle1_100000.h5 is recorded on 2024-02-02
import re

def get_date_from_h5(h5_filenames):
    """Extracts the recording date from h5 filenames in the format
       camera12024-02-02T12_19_31DLC_resnet50_seq_8_keypoints_thethredFeb27shuffle1_100000.h5

    Args:
        h5_filenames: A single h5 filename (str) or a list of filenames (list).

    Returns:
        If a single filename is provided: The date as a string in the format YYYY-MM-DD
        If a list of filenames is provided: A list of corresponding dates.  
    """

    date_pattern = r"\d{4}-\d{2}-\d{2}"  # Pattern for YYYY-MM-DD

    if isinstance(h5_filenames, str):  # Handle a single filename
        match = re.search(date_pattern, h5_filenames)
        if match:
            return match.group(0)
        else:
            return None  # Date not found in filename

    elif isinstance(h5_filenames, list):  # Handle a list of filenames
        dates = []
        for filename in h5_filenames:
            match = re.search(date_pattern, filename)
            if match:
                dates.append(match.group(0))
            else:
                dates.append(None)  # Indicate date not found
        return dates

    else:
        raise TypeError("Input must be a string or a list of strings.")


    
# h5_filename = 'camera12024-02-02T12_19_31DLC_resnet50_seq_8_keypoints_thethredFeb27shuffle1_100000.h5'

# print(get_date_from_h5(h5_filename))
```

```python
def get_deeplabcut_filename(h5_files, dates):
    '''
    Find and return a list of H5 files that match the specified date pattern.

    Args:
        h5_files (list): List of H5 files.
        dates (str or list): Date or list of dates to match.

    Returns:
        list: List of H5 files that match any of the date patterns, without duplicates.
    '''
    if isinstance(dates, str):
        dates = [dates]  # Convert to list for uniform processing

    matching_files = []

    for date in dates:
        date_pattern = fr"camera1{date}T\d{{2}}_\d{{2}}_\d{{2}}DLC_resnet50_seq_8_keypoints_thethredFeb27shuffle1_100000\.h5"
        for filename in h5_files:
            if re.match(date_pattern, filename) and filename not in matching_files:
                matching_files.append(filename)

    return matching_files


def get_timestamps_filename(timestamps_files, date):

    date_pattern = r"stamps" + date + r"T\d{2}_\d{2}_\d{2}\.csv"  
    for filename in timestamps_files:
        match = re.match(date_pattern, filename)
        if match:
            return match.group(0)  # Return the full matching filename

    return None  # No match found
```

```python
def get_date_from_h5(h5_filenames):
    """Extracts the recording date from h5 filenames in the format
       camera12024-02-02T12_19_31DLC_resnet50_seq_8_keypoints_thethredFeb27shuffle1_100000.h5

    Args:
        h5_filenames: A single h5 filename (str) or a list of filenames (list).

    Returns:
        If a single filename is provided: The date as a string in the format YYYY-MM-DD.
        If a list of filenames is provided: A list of corresponding dates without None values.  
    """

    date_pattern = r"\d{4}-\d{2}-\d{2}"  # Pattern for YYYY-MM-DD

    if isinstance(h5_filenames, str):  # Handle a single filename
        match = re.search(date_pattern, h5_filenames)
        return match.group(0) if match else None  # Date not found in filename

    elif isinstance(h5_filenames, list):  # Handle a list of filenames
        dates = [re.search(date_pattern, filename).group(0) for filename in h5_filenames if re.search(date_pattern, filename)]
        return dates

    else:
        raise TypeError("Input must be a string or a list of strings.")

    
def get_date_from_csv(csv_filenames):
    """Extracts the recording date from csv filenames in the format
        stamps2024-02-02T12_19_31.csv

    Args:
        csv_filenames: A single csv filename (str) or a list of filenames (list).

    Returns:
        If a single filename is provided: The date as a string in the format YYYY-MM-DD
        If a list of filenames is provided: A list of corresponding dates.  
    """

    date_pattern = r"\d{4}-\d{2}-\d{2}"

    if isinstance(csv_filenames, str):  # Handle a single filename
        match = re.search(date_pattern, csv_filenames)
        if match:
            return match.group(0)
        else:
            return None  # Date not found in filename

    elif isinstance(csv_filenames, list):  # Handle a list of filenames
        dates = []
        for filename in csv_filenames:
            match = re.search(date_pattern, filename)
            if match:
                dates.append(match.group(0))
            else:
                dates.append(None)  # Indicate date not found
        return dates

    else:
        raise TypeError("Input must be a string or a list of strings.")
    
```

```python
def get_all_xy_keypoints(animals, camera_folder, allsessions_df, opto_session_dates, get_date_from_h5, get_timestamps_filename, get_deeplabcut_filename, convert_and_uncycle_timestamps):
    """
    Generates a dataframe containing x, y keypoints for all specified animals and their trials.

    Parameters:
    - animals: A list of animal IDs.
    - camera_folder: Base path to the camera data.
    - allsessions_df: Dataframe containing all session data.
    - opto_session_dates: Dictionary mapping animal IDs to their opto session dates.
    - get_date_from_h5: Function to extract dates from h5 filenames.
    - get_timestamps_filename: Function to get the filename of timestamps.
    - get_deeplabcut_filename: Function to get the filename of deeplabcut predictions.

    Returns:
    - A pandas dataframe with x, y keypoints, trial_id, animal_id, and date for each trial.
    """
    # List to store all the data, later convert to a dataframe
    allxy_df = []

    # Loop through each animal
    for current_animal in animals:
        print(f"analysing animal: {current_animal}")
        df_current_animal = allsessions_df[allsessions_df['animal_id'] == current_animal]

        timestamps_folder = os.path.join(camera_folder, current_animal)
        deeplabcut_folder = os.path.join(camera_folder, current_animal, "predictions")

        # Get the dates to analyse for the animal
        opto_dates = opto_session_dates[current_animal]

        # Get the unique deeplabcut folders for the animal if exists
        h5_files = os.listdir(deeplabcut_folder)
        h5_dates = get_date_from_h5(h5_files)

        timestamps_files = [f for f in os.listdir(timestamps_folder) if f.endswith('.csv')]
        timestamps_files = [get_timestamps_filename(timestamps_files, date) for date in h5_dates]
        dlc_files = [get_deeplabcut_filename(h5_files, date) for date in h5_dates]
        dlc_files = [item for sublist in dlc_files for item in sublist]

        opto_dates_str = [date.strftime('%Y-%m-%d') for date in opto_dates]
        h5_dates_str = list(set([str(date) for date in h5_dates]))
        dates_to_analyse = list(set(opto_dates_str) & set(h5_dates_str))

        for date in dates_to_analyse:
            print(f"analysing date: {date}")

            # Get the behaviour data for df_current_animal on date
            behaviour_df = df_current_animal[df_current_animal['date'] == date]

            # Get all non-stimulated trials when behaviour_df['opto_condition is not 1
            # filtered_df = behaviour_df[behaviour_df['opto_condition'] != 1]
            filtered_df = behaviour_df.copy()

            # Get timestamps files for the date
            timestamps_file = get_timestamps_filename(timestamps_files, date)
            timestamps_path = os.path.join(timestamps_folder, timestamps_file)
            timestamps_df = pd.read_csv(timestamps_path, delimiter=' ', header=None)
            timestamps_df = timestamps_df.drop(columns=[3])
            timestamps_df.columns = ['pinstate', 'timestamp', 'frame_number']
            timestamps_df['pin_state_bool'] = timestamps_df['pinstate'] != timestamps_df['pinstate'][0]

            # Convert and uncycle timestamps
            timestamp_array = timestamps_df['timestamp'].values
            timestamps_in_seconds_uncycled = convert_and_uncycle_timestamps(timestamp_array)
            timestamps_df['time_seconds'] = timestamps_in_seconds_uncycled
            timestamps_df['frame_number'] = timestamps_df['frame_number'] - timestamps_df['frame_number'].iloc[0]

            # Get the deeplabcut file for the date
            # debug list index out of range
            # if get_deeplabcut_filename(dlc_files, date) returns an empty list or none, then skip the date
            if not get_deeplabcut_filename(dlc_files, date):
                print(f"no deeplabcut file found for date: {date}, skipping")
                continue
            dlc_file = get_deeplabcut_filename(dlc_files, date)[0]
            dlc_path = os.path.join(deeplabcut_folder, dlc_file)
            dlc_df = pd.read_hdf(dlc_path)
            dlc_df.columns = dlc_df.columns.droplevel(0)

            # Loop through unique trials in filtered_df and get the data from dlc_df and timestamps_df
            start_port = 2
            unique_trials = filtered_df['cumulative_trial_id'].to_list()

            for index, trial_id in enumerate(unique_trials):

                # print(f"analysing trial: {trial_id}") # debug

                # Check the length of the trial, if its 1 or less, skip
                if len(filtered_df[filtered_df['cumulative_trial_id'] == trial_id]) <= 1:
                    print(f"trial: {trial_id} is empty, skipping")
                    continue

                trial_df = filtered_df[filtered_df['cumulative_trial_id'] == trial_id]
                trial_start_index = trial_df[trial_df['start_poke_port'] == start_port].index.min()
                trimmed_trial_df = trial_df.loc[trial_start_index:]

                # Check if trimmed_trial_df is empty, if yes, skip
                if trimmed_trial_df.empty:
                    print(f"trial: {trial_id} is empty, skipping")
                    continue

                camera_start = trimmed_trial_df['camera_trial_start_times'].values[0]
                camera_end = trimmed_trial_df['camera_trial_end_times'].values[0]

                # If trial_id + 1 is not an index in behaviour_df, then the trial is the last trial, skip it
                if trial_id + 1 not in behaviour_df['cumulative_trial_id'].values:
                    print(f"trial: {trial_id} is the last trial, skipping")
                    continue

                # Find the next trial start time which is the next cumulative_trial_id in behaviour_df
                next_trial_camera_start = behaviour_df[behaviour_df['cumulative_trial_id'] == trial_id + 1]['camera_trial_start_times'].values[0]

                trial_timestamps = timestamps_df[(timestamps_df['time_seconds'] >= camera_start) & (timestamps_df['time_seconds'] <= next_trial_camera_start)]

                # Get frames to exctract
                frames_to_extract = trial_timestamps['frame_number'].values

                # Exctract these frames from dlc_df
                trial_df_keys = dlc_df.loc[frames_to_extract]

                # Get the x, y co-ordinates for the port times
                x_nose = trial_df_keys['nose', 'x']
                y_nose = trial_df_keys['nose', 'y']

                x_left_ear = trial_df_keys['left_ear', 'x']
                y_left_ear = trial_df_keys['left_ear', 'y']

                x_right_ear = trial_df_keys['right_ear', 'x']
                y_right_ear = trial_df_keys['right_ear', 'y']

                x_data = (x_nose + x_left_ear + x_right_ear) / 3
                y_data = (y_nose + y_left_ear + y_right_ear) / 3

                # Append the data to allxy_df
                allxy_df.append(pd.DataFrame({'x': x_data, 'y': y_data, 'trial_id': trial_id, 'animal_id': current_animal, 'date': date}))

    # Convert allxy_df to a dataframe
    allxy_df = pd.concat(allxy_df)

    return allxy_df

# get the data
allxy_df = get_all_xy_keypoints(Animals, Camera_Folder, allsessions_DF, opto_session_dates, get_date_from_h5, get_timestamps_filename, get_deeplabcut_filename, convert_and_uncycle_timestamps)
allxy_df.head()
```
